```python
# Importar librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 1. Cargar el dataset de cáncer de mama
data = load_breast_cancer()
X = data.data
y = data.target
print(X.shape, y.shape)

# 2. Dividir el dataset en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# 3. Crear el modelo de red neuronal secuencial
model_3 = Sequential()
model_3.add(Dense(units=16, input_dim= X.shape[1], activation='sigmoid'))
model_3.add(Dense(units=8, activation='sigmoid'))
model_3.add(Dense(units=1, activation='sigmoid'))

# 4. Compilar el modelo
# - Optimizador: 'sgd' (Stochastic Gradient Descent)
# - Pérdida: 'binary_crossentropy' (clasificación binaria)
# - Métrica: precisión
model_3.compile(optimizer='sgd', loss='binary_crossentropy',
metrics=['accuracy'])

# 5. Entrenar el modelo
# - Número de épocas: 50
# - Tamaño del batch: tamaño de lote para actualizar los pesos (32 muestras por paso)
# - Validación: cconjunto de validación para monitorear el rendimiento
history = model_3.fit(X_train, y_train, epochs=50, batch_size=32,
                    validation_data=(X_test, y_test)
)

# 6. Evaluar el modelo con los datos de prueba
loss, accuracy = model_3.evaluate(X_test, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")

import matplotlib.pyplot as plt

# 7. Extraer historial de entrenamiento para graficar desempeño
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

# 8. Graficar curvas de pérdida y precisión

# Pérdida (loss)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Precisión (accuracy)
plt.subplot(1, 2, 2)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```
